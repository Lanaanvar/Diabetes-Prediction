```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
file_path = "diabetes.csv"
df = pd.read_csv(file_path)
print(df)
```

```python
df.shape
df.describe
df.iloc[:, 8].value_counts()
```

```python
df.groupby(df.iloc[:, 8]).mean()
```

```python
# X=df.drop(columns="Outcome", axis=1)
# Y=df["Outcome"]

# Assuming you want to drop the column at index 2
column_index_to_drop = 8
X = df.drop(df.columns[column_index_to_drop], axis=1)

# If you also want to set the dropped column as your target variable
Y = df.iloc[:, column_index_to_drop]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_predictions=model.predict(X_train)
accuracy=accuracy_score(X_train_predictions, Y_train)
print("Accuracy : ",accuracy)
```

```python
X_test_predictions=model.predict(X_test)
test_accuracy=accuracy_score(X_test_predictions, Y_test)
print("Accuracy : ",test_accuracy)
```
